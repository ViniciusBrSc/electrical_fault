```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df_original = pd.read_csv('../../data/raw/detection_data.csv')
df_original = df_original.drop(['Unnamed: 7','Unnamed: 8'], axis=1)
df_original = df_original.rename(columns={'Output (S)':'output'})
df = df_original.copy()
df
```

```python
X = df.drop('output', axis=1)
Y = df['output']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
X_train_ss = scaler.fit_transform(X_train)
X_test_ss = scaler.transform(X_test)
X_train_ss = pd.DataFrame(X_train_ss, columns=X_train.columns)
X_test_ss = pd.DataFrame(X_test_ss, columns=X_test.columns)
```

```python
df_train = pd.concat([Y_train.reset_index(), X_train_ss], axis=1)
df_test = pd.concat([Y_test.reset_index(), X_test_ss], axis=1)
```

```python
display(f'Train: {df_train.shape}', df_train.head(), f'Test: {df_test.shape}', df_test.head())
```
